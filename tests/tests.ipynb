# Test para probar los modulos de la Poke API

```python
import requests
import json
import random
```

```python
POKE_API_URL = "https://pokeapi.co/api/v2/"
```

### Obtener el tipo de pokemon según el nombre

```python
response = requests.get(POKE_API_URL+"pokemon/"+"ivysaur")
if response.status_code == 200:
    # Convertir la respuesta a formato JSON (si la API devuelve JSON)
    data = response.json()
    print(json.dumps(data["types"],indent=4))
else:
    print(f"Error en la solicitud: {response.status_code}")
```

```python
list_pokemon_types = []

for type_by_pokemon in data["types"]:
    print(type_by_pokemon["type"]["name"])
    list_pokemon_types.append(type_by_pokemon["type"]["name"])
```

```python
def get_pokemon_types(pokemon_name):
    response = requests.get(POKE_API_URL+"pokemon/"+pokemon_name)
    if response.status_code == 200:
        # Convertir la respuesta a formato JSON (si la API devuelve JSON)
        data = response.json()
        list_pokemon_types = []

        for type_by_pokemon in data["types"]:
            list_pokemon_types.append(type_by_pokemon["type"]["name"])
        return list_pokemon_types

    else:
        print(f"Error en la solicitud: {response.status_code}")
```

```python
get_pokemon_types("pikachu")
```

### Obtener un pokemon aleatorio según el tipo

```python
response = requests.get(POKE_API_URL+"type/"+"poison")
if response.status_code == 200:
    # Convertir la respuesta a formato JSON (si la API devuelve JSON)
    data = response.json()
    print(json.dumps(data["pokemon"],indent=4))
else:
    print(f"Error en la solicitud: {response.status_code}")
```

```python
list_pokemon_name = []

for pokemon_by_type in data["pokemon"][:10]:
    print(pokemon_by_type["pokemon"]["name"])
    list_pokemon_name.append(pokemon_by_type["pokemon"]["name"])
```

```python
pokemon_random = random.choice(list_pokemon_name)
print(pokemon_random)
```

```python
def get_pokemon_name(pokemon_type):
    response = requests.get(POKE_API_URL+"type/"+pokemon_type)
    if response.status_code == 200:
        # Convertir la respuesta a formato JSON (si la API devuelve JSON)
        data = response.json()
        list_pokemon_name = []

        for pokemon_by_type in data["pokemon"]:
            list_pokemon_name.append(pokemon_by_type["pokemon"]["name"])
        pokemon_random = random.choice(list_pokemon_name)
        return pokemon_random
    else:
        print(f"Error en la solicitud: {response.status_code}")
```

```python
get_pokemon_name("bug")
```

### Obtener el pokemon con nombre más largo según el tipo

```python
pokemon_max_name = max (list_pokemon_name, key=len)
print(pokemon_max_name)
```

```python
def get_pokemon_max_name(pokemon_type):
    response = requests.get(POKE_API_URL+"type/"+pokemon_type)
    if response.status_code == 200:
        # Convertir la respuesta a formato JSON (si la API devuelve JSON)
        data = response.json()
        list_pokemon_name = []

        for pokemon_by_type in data["pokemon"]:
            list_pokemon_name.append(pokemon_by_type["pokemon"]["name"])
        pokemon_max_name = max (list_pokemon_name, key=len)
        return pokemon_max_name
    else:
        print(f"Error en la solicitud: {response.status_code}")
```

```python
get_pokemon_max_name("electric")
```
